```python
def assign_code_cards(description):
    #description.astype(str)
    #description.lower()
    if pd.isna(description):
        return 0
    elif 'yellow' in description:
        return 1
    elif 'red card' in description:
        return 2
    else:
        return 0
```

### Transformation of subs 

```python
def assign_code_subs(description):
    if pd.isna(description):
        return 0
    elif 'Not reported' in description:
        return 1
    elif 'Tactical' in description:
        return 2
    elif 'Delay' in description:
        return 3
    elif 'Resting' in description:
        return 4
    elif 'Injury' in description:
        return 5
    elif 'Risk of booking' in description:
        return 6
    elif 'Special achievements' in description:
        return 7
    elif 'Substitution without replacement' in description:
        return 8
    else:
        return 0 
```

```python
df_game_event_output['subs_descFK'] = df_game_event_output['description'].loc[df_game_event_output['type_Substitutions']==1].apply(assign_code_subs).astype(int)
df_game_event_output['subs_descFK'] = df_game_event_output['subs_descFK'].fillna(0).astype(int)
```

```python
counter = 0
goal_desc_set = set()
for i in df_game_event_output['description'].loc[df_game_event_output['type_Goals']==1].str.split(','):
    for y in i :
        goal_desc_set.add(y)
print(goal_desc_set)
```

```python
goal_desc_set_filter = set()
keywords = ['Season', 'Tournament', 'Assist', 'assist']
for desc in goal_desc_set : 
    if any(keyword in desc for keyword in keywords):
        continue
    elif desc == '':
        continue
    else :
        goal_desc_set_filter.add(desc)
```

```python
goal_dim = np.array([[0,'No Goal']])
```

```python
id = 1 
for goal_desc in goal_desc_set_filter:
    goal_dim = np.append(goal_dim,[[id,goal_desc]],axis=0)
    id +=1
goal_dim
```

```python
def assign_code_goal(description, goal_dim) : 
    if pd.isna(description):
        return 0
    else: 
        for index , desc in goal_dim:
            if str(desc) in description:
                return index
        else:
            return -1
```

```python
df_game_event_output['goal_descFK'] = df_game_event_output['description'].loc[df_game_event_output['type_Goals']==1].apply(lambda description: assign_code_goal(description, goal_dim)).astype(int)
df_game_event_output['goal_descFK'] = df_game_event_output['goal_descFK'].fillna(0).astype(int)
```

### Assist ETL 

```python
df_assist_desc =df_game_event_output['description'].loc[
    (df_game_event_output['type_Goals'] == 1) & 
    (df_game_event_output['description'].str.contains('assist',case = False ,na=False))
]
df_assist_desc
```

```python
assist_set = set()
for i in df_assist_desc.str.split(','):
    for y in i : 
        assist_set.add(y)
```

```python
assist_set
```

```python
assist_desc_set_filter = set()
keywords = ['Season', 'Tournament']
for desc in assist_set : 
    if any(keyword in desc for keyword in keywords):
        continue
    elif desc == '':
        continue
    else :
       assist_desc_set_filter.add(desc)
```

```python
assist_desc_set_filter
```

```python
assist_desc_set_filter = {'Pass','Corner','Cross','Throw-in','Free kick','Without assist'}
```

```python
assist_dim = np.array([
    [0, 'No Assist'],
    [1, 'Pass'],
    [2, 'Corner'],
    [3, 'Throw-in'],
    [4, 'Free kick']
])
```

```python
def assign_code_assist(description):
    if pd.isna(description):
        return 0
    elif 'Pass' in description:
        return 1
    elif 'Corner' in description:
        return 2
    elif 'Throw-in' in description:
        return 3
    elif 'Free kick' in description:
        return 4
    else:
        return 0 
```

```python
df_game_event_output['assist_descFK'] = df_game_event_output['description'].loc[df_game_event_output['type_Goals']==1].apply(assign_code_assist).astype(int)
df_game_event_output['assist_descFK'] = df_game_event_output['assist_descFK'].fillna(0).astype(int)
```
